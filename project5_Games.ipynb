## Подготовка данных

Исправим заголовки столбцов и проведём трансформации типов данных.

Приведём названия столбцов в соответствии с хорошим стилем к нижнему регистру. Для этого заменим заглавные буквы на строчные.

```python
# переименование столбцов
df = df.rename(columns = {'Name':'name', 'Platform':'platform',\
                          'Year_of_Release':'year_of_release', 'Genre':'genre',\
                          'NA_sales':'na_sales', 'EU_sales':'eu_sales', 'JP_sales':'jp_sales',\
                          'Other_sales':'other_sales', 'Critic_Score':'critic_score',\
                          'User_Score':'user_score', 'Rating':'rating'})
```

Проверим результат. Для этого ещё раз выведем на экран названия столбцов:

```python
# перечень названий столбцов таблицы df
df.columns
```

Перед изменением типов данных необходимо избавиться от пропусков.

Подсчитаем долю пропусков в различных столбцах.

```python
# подсчёт пропусков
df.isna().mean()
```

Выведем на экран строки с пропусками в столбцах.

```python
# получение строк таблицы с пропущенными значениями в столбце name
df[df['name'].isna()].head()
```

В этих строках одновременно отсутствуют данные в пяти столбцах. На что их менять, непонятно. Доля пропусков 0,012%. По совокупности причин эти строки целесообразнее удалить.

```python
# удаление пропущенных значений в столбце name
df = df.dropna(subset=['name'])
```

```python
# получение 25 строк таблицы с пропущенными значениями в столбце year_of_release
df[df['year_of_release'].isna()].head(25)
```

Пропусков в столбце `year_of_release` относительно много (1,6%) и многие из этих игр имеют значительные продажи. 

Однако данные о годе релиза игры - ключевой параметр для анализа и его отсутствие делает данные малопригодными для дальнейшего исследования. Кроме того, удаление 267 строк может сместить оценки по жанрам игр, платформам и оценкам игр критиками и пользователями.

Поэтому лучше удалить пропуски.

```python
# удаление пропущенных значений в столбце year_of_release
df = df.dropna(subset=['year_of_release'])
```

Осталось решить, что делать с пропущенными значениями оценок критиков, пользователей и рейтинга ESRB - таких значений от 40% до 51%. Такое количество пропусков могло возникнуть из-за отсутствия статистики за ранние периоды или на отдельных платформах, из-за нежелания пользователей и критиков тратить время на выставление оценок (в случае малораспространённых игр), из-за технических сбоев либо по иным причинам.

Удаление такого массивы данных крайне негативно скажется на анализе и моделировании прогноза, поэтому заменим пропуски в этих столбцах на заглушку `-1`.

```python
# замена пропущенных значений в столбцах critic_score, user_score, rating на -1
df['critic_score'] = df['critic_score'].fillna(-1)
df['user_score'] = df['user_score'].fillna(-1)
df['rating'] = df['rating'].fillna(-1)
```

Как мы видели в выборке нулевых значений столбца `year_of_release`, в столбце `user_score` есть ещё странное значение `tbd`. Оно означает `To Be Determined`, то есть значение будет определено в дальнейшем. Возможно, ещё слишком малое число пользователей проставили оценки, либо ещё идёт агрегирование оценок от разных источников, либо к этому привели другие причины.

По сути, значение `tbd` равнозначно `NaN`. Поэтому также заменим их на `-1`.

```python
# замена значений tbd в столбце user_score на -1
df['user_score'] = df['user_score'].replace(['tbd'], -1)
```

Проверим теперь таблицу на наличие пропусков.

```python
# подсчёт пропусков
df.isna().sum()
```

Теперь пропусков нет, можем менять типы данных.

```python
# трансформация типов данных
df['year_of_release'] = df['year_of_release'].astype('int')
df['na_sales'] = df['na_sales'].astype('float32')
df['eu_sales'] = df['eu_sales'].astype('float32')
df['jp_sales'] = df['jp_sales'].astype('float32')
df['other_sales'] = df['other_sales'].astype('float32')
df['critic_score'] = df['critic_score'].astype('float32')
df['user_score'] = df['user_score'].astype('float32')
```

## Проверка гипотез

Проверим две гипотезы:

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза: пользовательский рейтинг платформы XOne равен пользовательскому рейтингу платформы PC.
Альтернативная гипотеза: пользовательский рейтинг платформы XOne не равен пользовательскому рейтингу платформы PC.

Рейтинги платформ не зависят друг от друга, разброс значений совокупностей примерно одинаков.

```python
# проверка нулевой гипотезы о равенстве средних двух независимых генеральных
# совокупностей xone и pc
xone = df_actual.query('platform == "XOne" and user_score > 0')['user_score']
pc = df_actual.query('platform == "PC" and user_score > 0')['user_score']
alpha = 0.05
results = st.ttest_ind(xone, pc)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем H0")
else:
    print("Не получилось отвергнуть H0")
```

Проверка показала, что пользовательский рейтинг платформы XOne равен пользовательскому рейтингу платформы PC. Различие их пользовательских рейтингов статистически незначимо и случайно.

Проверим вторую гипотезу.

Нулевая гипотеза: пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») статистически значимо отличаются друг от друга.

Альтернативная гипотеза состоит в том, что пользовательские рейтинги жанров Action и Sports равны.

Генеральные совокупности не зависят друг от друга.

```python
# проверка нулевой гипотезы о равенстве средних двух независимых генеральных
# совокупностей action и sports
action = df_actual.query('genre == "Action" and user_score > 0')['user_score']
sports = df_actual.query('genre == "Sports" and user_score > 0')['user_score']
alpha = 0.05
results = st.ttest_ind(action, sports)
print('p-значение:', results.pvalue)
if results.pvalue >= alpha:
    print("Отвергаем H0")
else:
    print("Не получилось отвергнуть H0")
```

Проверка показала, что средний пользовательский рейтинг игры жанра Action статистически значимо отличается от пользовательского рейтинга игры жанра Sports.

Итак, исследование завершено. Сформулируем его общий вывод.
